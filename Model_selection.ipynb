```python
# prompt: preprocess data using standard scalar on X

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
## fit ==> Mu mean, std standdeviation
## transform   z_score = (X - mu)/std.
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
# prompt: train a logistic regression model where X is  X_train and y is y_train, then evaluate it on X_test where ground truth is y_test

from sklearn.linear_model import LogisticRegression
def logisitic_regression_hands_on(X_train,y_train,X_test,y_test):
  model = LogisticRegression()
  model.fit(X_train,y_train)
  y_pred = model.predict(X_test)

  accuracy = accuracy_score(y_test,y_pred)
  report = classification_report(y_true=y_test, y_pred=y_pred)
  print('Logistic regression')
  print(f' Accuracy of logistic regression is {accuracy}')
  print('Classification report \n',report)
# model.score(X_test,y_test)
```

```python
logisitic_regression_hands_on(X_train,y_train,X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def KNN_classifier_hands_on(X_train,y_train,X_test,y_test):
  model  = KNeighborsClassifier(n_neighbors=3)
  model.fit(X_train,y_train)
  y_pred = model.predict(X_test)

  accuracy = accuracy_score(y_true=y_test,y_pred=y_pred)
  report = classification_report(y_true=y_test, y_pred=y_pred)
  print(' K Neighbor classifier')
  print(f' Accuracy of KNN is {accuracy}')
  print('Classification report \n',report)
```

```python
KNN_classifier_hands_on(X_train,y_train,X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
def decision_tree_hands_on(X_train,y_train,X_test,y_test):
  model = DecisionTreeClassifier()
  model.fit(X_train,y_train)
  y_pred =model.predict(X_test)

  ## calculate the accuracy and report
  accuracy = accuracy_score(y_test,y_pred)
  report = classification_report(y_test,y_pred)
  print(' Decision Tree')
  print(f'Accuracy of Decision tree is {accuracy}')
  print('Classification Report \n',report)
```

```python
decision_tree_hands_on(X_train,y_train,X_test,y_test)
```
